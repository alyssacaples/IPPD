```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import warnings
from IPython.display import Audio
```

```python
f = open("WINGBEAT.txt", "r")

numbers = []

for x in f:
    numbers.append(int(x.strip()))

signalData = np.array(numbers)
print(len(signalData))
```

```python
plt.plot(signalData)

plt.xlabel('Sample')

plt.ylabel('Amplitude')
```

```python
samplingFrequency = 600
plt.specgram(signalData,Fs=samplingFrequency)
plt.xlabel('Time')
plt.ylabel('Frequency')

#nyquist needs to be one half the sampling rate
```

```python
Audio(signalData, rate=samplingFrequency)
```
